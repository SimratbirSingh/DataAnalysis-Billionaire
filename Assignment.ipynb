## Methods
### Data Collection

A subset of data from Subset Data to Billionaires listed in Forbes stats is maintained publicly at [Kaggle](https://www.kaggle.com/datasets/princeiornongu/subset-data-to-billionaires-listed-in-forbes), a free online resource of general datasets [1].

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from IPython.display import display
```

```python
# Load data from local folder
billionaire = pd.read_csv('BillionairesForbes.csv')
display(billionaire)
```

### Data Cleaning

I changed the names of a couple of the Bilionaire's dataset's strangely named columns. The columns that are required for my analysis have been kept. Since no missing or NAN data have been observed, no data have been erased.

```python
# Keep desired columns
desired_columns = ['Rank', 'Forbes_ID', 'Net_Worth_($B)', 'Industry', 'Self-Made Score', 'Bachelor', 'Master', 'Doctorate', 'Drop Out']
billionaire = billionaire[desired_columns]

# Rename column names
billionaire.columns = ["rank", "id", "networth", "industry", "selfmade_score", "bachelor", "master", "doctorate", "drop out"]
display(billionaire)
```

### Distributions

The provided code calculates and prints descriptive statistics for two variables, namely net worth and self-made score, from a dataset of billionaires. The net worth statistics reveal the mean, maximum, and minimum values of billionaires' net worth. Similarly, the self-made score statistics provide the mean, maximum, and minimum values for a score reflecting the extent to which billionaires are self-made. These summary statistics offer a concise overview of the distribution and range of wealth and self-made success among the billionaires in the dataset.

As a reference, self-made score according to Forbes is measured as the following [1]:
- 1: Inherited fortune but not working to increase it
- 2: Inherited fortune and has a role in managing it
- 3: Inherited fortune and helping to increase it marginally
- 4: Inherited fortune and increasing it in a meaningful way
- 5: Inherited small or medium-size business and made it into a ten-digit fortune
- 6: Hired hand who didn’t create the business
- 7: Self-made who got a head start from wealthy parents and moneyed background
- 8: Self-made who came from a middle-class or upper-middle-class background
- 9: Self-made who came from a largely working-class background; rose from little to nothing
- 10: Self-made who not only grew up poor but also overcame significant obstacles

#### Education VS Self-Made Score

The bar plot visualizes the frequency distribution of self-made scores based on different education levels among billionaires. The majority of billionaires with Bachelor's degrees achieved self-made scores of 6 and 8, while those with Master's degrees generally scored 8, and individuals with Doctorates achieved scores of 6, 8, and 10. Dropouts exhibit a broad range of self-made scores, with a notable concentration around 6 and 8. Overall, the plot highlights the diverse educational backgrounds associated with varying levels of entrepreneurial success among billionaires.








```python
# Filter rows with Bachelor's degree
bachelor_df = billionaire[billionaire['bachelor'] == 1]

# Filter rows with Master's degree
master_df = billionaire[billionaire['master'] == 1]

# Filter rows with Doctorate
doctorate_df = billionaire[billionaire['doctorate'] == 1]

# Filter rows with Drop Out
dropout_df = billionaire[billionaire['drop out'] == 1]

# Group by self-made score and calculate the frequency for Bachelor's degree
bachelor_grouped = bachelor_df.groupby('selfmade_score').size().reset_index(name='Bachelor Frequency')

# Group by self-made score and calculate the frequency for Master's degree
master_grouped = master_df.groupby('selfmade_score').size().reset_index(name='Master Frequency')

# Group by self-made score and calculate the frequency for Doctorate
doctorate_grouped = doctorate_df.groupby('selfmade_score').size().reset_index(name='Doctorate Frequency')

# Group by self-made score and calculate the frequency for Drop Out
dropout_grouped = dropout_df.groupby('selfmade_score').size().reset_index(name='Drop Out Frequency')

# Plot the bar chart
plt.figure(figsize=(12, 8))

# Plot bars for Bachelor's degree
plt.bar(bachelor_grouped['selfmade_score'], bachelor_grouped['Bachelor Frequency'], color='blue', label='Bachelor')

# Plot bars for Master's degree
plt.bar(master_grouped['selfmade_score'], master_grouped['Master Frequency'], color='purple', label='Master', alpha=0.7)

# Plot bars for Doctorate
plt.bar(doctorate_grouped['selfmade_score'], doctorate_grouped['Doctorate Frequency'], color='pink', label='Doctorate', alpha=0.7)

# Plot bars for Drop Out
plt.bar(dropout_grouped['selfmade_score'], dropout_grouped['Drop Out Frequency'], color='orange', label='Drop Out', alpha=0.7)

plt.title('Frequency of Self-Made Score for Different Education Levels')
plt.xlabel('Self-Made Score')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

#### Industry VS Self-Made Score

The code processes data on billionaires, grouping them by industry and self-made score to calculate the frequency of each score. It then generates a bar plot illustrating the distribution of self-made scores across various industries. The resulting visualization provides a clear comparison of self-made scores, showcasing the distribution patterns and frequency within each industry.








```python
# Group by industry and self-made score and calculate the frequency
grouped_df = billionaire.groupby(['industry', 'selfmade_score']).size().reset_index(name='Frequency')

# Plot the bar chart
plt.figure(figsize=(12, 8))

# Get unique industries for x-axis ticks
industries = grouped_df['industry'].unique()

# Set up color map for bars
colors = plt.cm.get_cmap('tab10', len(industries))

# Plot bars for each industry
for i, industry in enumerate(industries):
    industry_data = grouped_df[grouped_df['industry'] == industry]
    plt.bar(industry_data['selfmade_score'], industry_data['Frequency'], label=industry, color=colors(i), alpha=0.7)

plt.title('Frequency of Self-Made Score for Different Industries')
plt.xlabel('Self-Made Score')
plt.ylabel('Frequency')
plt.legend(title='Industry', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
